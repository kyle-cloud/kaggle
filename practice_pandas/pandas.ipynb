```python
import pandas as pd
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                    'key2': ['K0', 'K1', 'K0', 'K1'],
                    'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3']})
```

```python
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                    'key2': ['K0', 'K0', 'K0', 'K0'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
res = pd.merge(left, right, on = 'key')
```

```python
res
```

```python
left
```

```python
right
```

```python
res = pd.merge(left, right, on = ['key1', 'key2'], left_index = True, right_index = True, how = 'outer', indicator = True)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.Series(np.random.randn(1000), index = np.arange(1000))
data = data.cumsum()
```

```python
data.plot()
plt.show()
```

```python
data = pd.DataFrame(np.random.randn(1000, 4), index = np.arange(1000), columns = list("ABCD"))
```

```python
data.head()
```

```python
data = data.cumsum()
data.plot()
plt.show()
```

```python
ax = data.plot.scatter(x = 'A', y = 'B', color = 'DarkBlue', label = 'Class 1')
data.plot.scatter(x = 'A', y = 'C', color = 'DarkGreen', label = 'Class 2', ax = ax)
plt.show()
```
